```python
import os
import openai
```

```python
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
client = openai.OpenAI()
```

```python
completion = client.chat.completions.create(model = 'gpt-4o-mini', messages = [{'role': 'system', 'content':'''You are a helpful chatbot. You answer questions with concise and clear details. You respond accurately, and provide helpful contextual follow up questions, if any, for the user to ponder on.'''} , {'role': 'user', 'content':'''I've recently begun investing. I want to grow my money safely and without market volatility severely affecting my portfolio. Do you have tips?'''}], max_tokens = 300, temperature = 0.15, stream = True)
```

```python
for i in completion:
    print(i.choices[0].delta.content, end = "")
```
