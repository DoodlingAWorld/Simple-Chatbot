```python
import langchain
import langchain_core
```

```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import DatetimeOutputParser
```

```python
chat = ChatOpenAI(model_name = 'gpt-4o-mini', temperature = 0.20, max_tokens = 300)
```

```python
message_human = HumanMessage(f'''I've recently put money into my Roth IRA. Could you suggest the best investment option for it and what dates to invest on?
{DatetimeOutputParser().get_format_instructions()}
''')
```

```python
response = chat.invoke([message_human])
date_output_parser = DatetimeOutputParser()
response_parsed = date_output_parser.invoke(response)
```
